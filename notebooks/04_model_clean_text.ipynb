# 04_model_clean_text.ipynb

This notebook trains a baseline logistic regression model using only the cleaned event description stored in "clean_description" (lemmatized) from the ACLED dataset.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load preprocessed dataset (with clean_description)
df = pd.read_pickle("acled_cleaned.pkl")
```

```python
# 2. Filter to necessary columns
X_text = df['clean_description']
y = df['event_type']

```

```python
# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_text, y, test_size=0.2, stratify=y, random_state=42
)

```

```python
# 4. TF-IDF vectorization
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# 5. Train logistic regression model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train_tfidf, y_train)


```

```python
# 6. Predict and evaluate
y_pred = clf.predict(X_test_tfidf)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, target_names=y.cat.categories))

```

```python
# 7. Confusion Matrix


# Generate normalized confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=y.cat.categories, normalize='true')

# Plot with seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=y.cat.categories,
            yticklabels=y.cat.categories)

plt.title("Normalized Confusion Matrix - Logistic Regression (Cleaned Text)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
# Optional interpretation hint
# A confusion matrix shows where the model confuses one class with another.
# Ideal model has strong diagonal (correct predictions) and few off-diagonal elements.
```
