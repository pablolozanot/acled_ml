```python
# 01_exploration.ipynb

import pandas as pd
from pandas_gbq import read_gbq
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Query ACLED data sample
query = """
SELECT 
  event_id_cnty,
  event_type,
  event_date,
  country,
  admin1,
  admin2,
  actor1,
  actor2,
  notes AS event_description,
  fatalities,
  latitude,
  longitude
FROM `my-acled-events.acled_events.acled_events`
WHERE event_type IS NOT NULL
  AND latitude IS NOT NULL
  AND longitude IS NOT NULL
LIMIT 20000
"""
```

```python
# A new browser tab will pop up for authentication through google
df = read_gbq(query, project_id="my-acled-events", use_bqstorage_api=True)
df.head()
```

```python
# Ensure fatalities is numeric
df['fatalities'] = pd.to_numeric(df['fatalities'], errors='coerce')
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
```

```python
df.to_csv('acled_sample.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='event_type', order=df['event_type'].value_counts().index, palette='Set2')
plt.title('Event Type Distribution')
plt.xlabel('Count')
plt.ylabel('Event Type')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Create the FacetGrid
g = sns.FacetGrid(
    df,
    col="event_type",
    col_wrap=3,
    height=3.5,
    sharex=False,
    sharey=False
)

g.map_dataframe(sns.histplot, x="fatalities", bins=30, color="skyblue")

# Tidy up
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Fatalities", "Count")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Fatalities Distribution by Event Type", fontsize=16)
plt.show()
```

```python
df_map = df.dropna(subset=['latitude', 'longitude'])

import folium
from folium.plugins import MarkerCluster

# Use a small sample for performance
sample_map = df_map.sample(n=1000, random_state=42)

# Center the map roughly around global data
m = folium.Map(location=[10, 0], zoom_start=2)

# Add marker clusters
marker_cluster = MarkerCluster().add_to(m)

for _, row in sample_map.iterrows():
    popup = f"<strong>{row['event_type']}</strong><br>{row['event_description'][:100]}..."
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=popup,
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(marker_cluster)

m
```

```python
m.save("acled_map.html")
```

```python
# Define a color map for event types
color_map = {
    'Battles': 'red',
    'Protests': 'blue',
    'Riots': 'orange',
    'Explosions/Remote violence': 'purple',
    'Strategic developments': 'green',
    'Violence against civilians': 'darkred'
}
import folium
from folium.plugins import MarkerCluster

# Sample and clean the data
sample_map = df_map.sample(n=1000, random_state=42)

# Create map
m = folium.Map(location=[10, 0], zoom_start=2)
marker_cluster = MarkerCluster().add_to(m)

for _, row in sample_map.iterrows():
    event_type = row['event_type']
    popup = f"<strong>{event_type}</strong><br>{row['event_description'][:100]}..."
    color = color_map.get(event_type, 'gray')

    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=popup
    ).add_to(marker_cluster)

m
```

```python
m.save("acled_map2.html")
```
